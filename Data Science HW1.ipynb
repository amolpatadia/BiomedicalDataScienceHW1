```python
from urllib.request import urlretrieve
```

```python
import numpy as np
import pandas as pd
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/postoperative-patient-data/post-operative.data'
```

```python
urlretrieve(url, 'post-operative.data')
```

```python
header = ['L-CORE','L-SURF','L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','ADM-DECS']
df = pd.read_csv('post-operative.data',names = header, comment = '#')
```

# Need to replace all missing values ('?') with 'NaN' so that it can be imputed

```python
df['COMFORT'] = df['COMFORT'].apply(pd.to_numeric, errors = 'coerce')
```

```python
df
```

# Now I am checking how many 'NaN' values are present and imputing them with the mean value so that I do not need to remove patient information and can still analyze those patients' data

```python
dfo = df
```

```python
df.isnull().sum()
```

```python
dfo.dropna()
```

I chose to utilize 2 preprocessing techniques. The first seen above is simply deleting the patient lines that have missing data. The second seen below is imputing the missing data with the mean. The first method has an issue in that we are getting rid of data points, but the second method has an issue in that it may not be reasonable to assume that the comfort of those patients is coordinated with the mean since it is subjective and may be based on other categorical data. For the plotting I decided to utilize the second method data to maximize the number of points that are displayed and to just keep in mind where the mean values lie in each distribution. 

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='mean')
df["COMFORT"] = imp.fit_transform(df[["COMFORT"]]).ravel()
```

```python
df
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# I used a swarmplot to plot each of the categories against COMFORT because it allowed me to see individual points as well as the overall distribution without all of the identical values overlapping. This way I could more easily observe any trends that are present. 

# L-Core vs Comfort

```python
sns.swarmplot(x=df['L-CORE'], y=df['COMFORT'])
```

The distribution of comfort based on the patients' internal temperatures all seem to have similar distribution. Most of each category had a comfort level around 10 with some at 15 and with 1 patient lower. The "high" group has the least drastic difference in the number of people rating 10 and 15. It appears that overall, there is not much difference in comfort level when comparing patients with different internal temperatures. 

# L-SURF vs COMFORT

```python
sns.swarmplot(x=df['L-SURF'], y=df['COMFORT'])
```

It appears that there is not much difference in the comfort ratings based on the patients' surface temperatures. Each group showed similar distributions with most ratings around 10 and a smaller cluster at 15. 

# L-O2 vs COMFORT

```python
sns.swarmplot(x=df['L-O2'], y=df['COMFORT'])
```

The distribution of comfort ratings based on oxygen saturation level was once again very similar between the patient groups. The "good' L-O2 group appears to have more ratings near 15 compared to the "excellent" group so it is possible that patients with L-O2 between 90% and 98% are a little more comfortable than those with higher L-O2 levels.

# L-BP vs COMFORT

```python
sns.swarmplot(x=df['L-BP'], y=df['COMFORT'])
```

Most of the patients in each of the blood pressure groups reported comfort levels near 10. The 'mid' L-BP group appears to have more patients reporting comfort levels of 15 so it seems that they are more comfortable than the patients with high or low blood pressures. 

# SURF-STBL vs COMFORT

```python
sns.swarmplot(x=df['SURF-STBL'], y=df['COMFORT'])
```

The distributions of patient comfort relative to stability of surface temperature are very similar. I do not believe there is a significant relationship between stability of surface temperature and patient comfort that is strong enough to see better or worse comfort levels in one group or another. 

# CORE-STBL vs COMFORT

```python
sns.swarmplot(x=df['CORE-STBL'], y=df['COMFORT'])
```

There is not enough data in the "mod-stable" group to properly interpret its plot. For the stable and unstable internal temperature groups, the distributions were differerent in that the stable group had much greater numbers of patients at the comfort 10 level and 15 level. The unstable group also had more of an even distribution between the comfort ratings of 10 and 15 but this may have been due to the lower number of patients in that category. Based on this plot,there is a greater proportion of unstable patients with higher comfort ratings implying that this group has a higher comfort level, but the unstable group did not have that many patients so I do not have full confidence in its plot being representative of that group. 

# BP-STBL vs COMFORT

```python
sns.swarmplot(x=df['BP-STBL'], y=df['COMFORT'])
```

The "stable" group seems to have a greater number of patients reporting comfort at 15, the "unstable" group has the second most reports of 15 comfort, and the "mod-stable" group has the least. All three groups have majority of their subjects reporting comfort levels of 10. However, the "mod-stable" and "unstable" groups each had a patient reporting a very low comfort score. It is hard to make any inferences in terms of which group is more comfortable but on average most are reporting levels near each other.
